```python
import numpy as np    
from scipy import stats
import matplotlib.pyplot as plt
```

## Variáveis Aleatórias Discretas - VAD

São variáveis $X: \Omega \rightarrow N$ cujo conjunto $N$ é um conjunto numerico finito ou infinito contável.

### Exemplo 1: Lançamento de Moeda

$$\Omega = \{K, C\}$$

$$X: \Omega \rightarrow \{0, 1\}$$

$$X \sim \mathcal{U}(\Omega)$$

```python
n = 50   # Número de Amostras (realizações de X)

x = np.random.randint(0,2, n)  # Gerador de números aleatórios inteiros

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=[10,3])

ax[0].plot(x,'ob')
ax[0].set_xlabel("Realização")
ax[0].set_ylabel("$\Omega$")

ax[1].hist(x, density=True)
ax[1].set_xlabel("$\Omega$")
ax[1].set_ylabel("Densidade")
```

### Exemplo 2: Lançamento de Um Dado
$$\Omega = \{1, 2, 3, 4, 5, 6\}$$

$$X: \Omega \rightarrow \mathbb{N}^+$$

$$X \sim \mathcal{U}(\Omega)$$

```python
n = 50000

x = np.random.randint(1,7, n)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=[10,3])

ax[0].plot(x,'ob')
ax[0].set_xlabel("Realização")
ax[0].set_ylabel("$\Omega$")

ax[1].hist(x, density=True)
ax[1].set_xlabel("$\Omega$")
ax[1].set_ylabel("Densidade")
```

## Variáveis Aleatórias Contínuas - VAC

### Exemplo 1: Ruído Branco

$$\Omega \in \mathbb{R}$$

$$ X: \Omega \rightarrow \mathbb{R}$$ 

$$X \sim \mathcal{N}(0,1)$$

```python
n = 3500
x = np.random.normal(0, 1, n)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=[10,3])

ax[0].plot(x,'ob')
ax[0].set_xlabel("Realização")
ax[0].set_ylabel("$\Omega$")

kde = stats.kde.gaussian_kde(x)
it = np.linspace(min(x), max(x), n)

ax[1].hist(x, density=True)
ax[1].plot(it, kde(it))
ax[1].set_xlabel("$\Omega$")
ax[1].set_ylabel("Densidade")
```

## Independência

### Duas VADs independentes entre si
$$A \sim \mathcal{U}(1,6)$$
$$B \sim \mathcal{U}(1,6)$$
Se $A$ e $B$ são independentes, então:
* $P(A,B) = P(A)\cdot P(B)$
* $P(A|B) = P(A)$
* $P(B|A) = P(B)$

```python
n = 10000
x1 = np.random.randint(1,7, n)
x2 = np.random.randint(1,7, n)

fig, ax = plt.subplots(nrows=1, ncols=4, figsize=[15,5])

ax[0].scatter(x1, x2)
ax[0].set_xlabel("$X_1$")
ax[0].set_ylabel("$X_2$")
ax[0].set_title("$X_1 \\times X_2$")

histograma_x1 = ax[1].hist(x1, bins=range(1,8), density=True)
ax[1].set_title("Histograma de $X_1$")
histograma_x2 = ax[2].hist(x2, bins=range(1,8), density=True)
ax[2].set_title("Histograma de $X_2$")

a = [k for k in zip(x1, x2)]
b = set(sorted(a))
index = {k:v for v,k in enumerate(b)}

histograma_x1x2 = ax[3].hist([index[k] for k in a], bins=range(len(b)+1), density=True)
ax[3].set_title("Histograma de $X_1 \\times X_2$")

plt.tight_layout()
```

```python
Px1 = { k: v for k,v in zip(histograma_x1[1],histograma_x1[0])}
Px2 = { k: v for k,v in zip(histograma_x2[1],histograma_x2[0])}
Px1x2 = { k: v for k,v in zip(histograma_x1x2[1],histograma_x1x2[0])}
```

```python
a = 1
b = 3
print('P(A) = {}'.format(Px1[a]))
print('P(B) = {}'.format(Px2[b]))
print('P(A) * P(B) = {}'.format(round(Px1[a]*Px2[b],3)))
print('P(A,B) = {}'.format(round(Px1x2[index[(a,b)]],3)))
print('P(A|B) = {}'.format(round(Px1x2[index[(a,b)]]/Px2[b],3)))
print('P(B|A) = {}'.format(round(Px1x2[index[(a,b)]]/Px1[a],3)))
```

### Duas VADs dependentes entre si

$$A \sim \mathcal{U}(1,6)$$
$$B = 2A + \mathcal{U}(1,6)$$

Se $A$ e $B$ não são independentes, então:
* $P(A,B) = P(A|B)\cdot P(B)$
* $P(A|B) = \frac{P(B|A)\cdot P(A)}{P(B)}$
* $P(B|A) = \frac{P(A|B)\cdot P(B)}{P(A)}$

```python
n = 5000
x1 = np.random.randint(1,7, n)
x2 = 2*x1 + np.random.randint(1,7, n)

fig, ax = plt.subplots(nrows=1, ncols=4, figsize=[15,5])

ax[0].scatter(x1, x2)
ax[0].set_xlabel("$X_1$")
ax[0].set_ylabel("$X_2$")
ax[0].set_title("$X_1 \\times X_2$")

histograma_x1 = ax[1].hist(x1, bins=range(1,8), density=True)
ax[1].set_title("Histograma de $X_1$")
histograma_x2 = ax[2].hist(x2, bins=range(min(x2),max(x2)+1), density=True)
ax[2].set_title("Histograma de $X_2$")

a = [k for k in zip(x1, x2)]
b = set(sorted(a))
index = {k:v for v,k in enumerate(b)}

histograma_x1x2 = ax[3].hist([index[k] for k in a], bins=range(len(b)+1), density=True)
ax[3].set_title("Histograma de $X_1 \\times X_2$")

plt.tight_layout()
```

```python
Px1 = { k: v for k,v in zip(histograma_x1[1],histograma_x1[0])}
Px2 = { k: v for k,v in zip(histograma_x2[1],histograma_x2[0])}
Px1x2 = { k: v for k,v in zip(histograma_x1x2[1],histograma_x1x2[0])}
```

```python
a = 1
b = 8
print('P(A) = {}'.format(Px1[a]))
print('P(B) = {}'.format(Px2[b]))
print('P(A) * P(B) = {}'.format(round(Px1[a]*Px2[b],3)))
print('P(A,B) = {}'.format(round(Px1x2[index[(a,b)]],3)))
print('P(A|B) = {}'.format(round(Px1x2[index[(a,b)]]/Px2[b],3)))
print('P(B|A) = {}'.format(round(Px1x2[index[(a,b)]]/Px1[a],3)))
```

## Expressões com Variáveis Aleatórias

### VAs Discretas: Soma do Lançamento de Dois Dados

$$\Omega = \{1, 2, 3, 4, 5, 6\}$$
$$X_1 \sim \mathcal{U}(\Omega)$$
$$X_2 \sim \mathcal{U}(\Omega)$$

$$ Y: \Omega \times \Omega \rightarrow \mathbb{N}⁺$$

$$Y = X_1 + X_2$$

```python
n = 50000

x1 = np.random.randint(1,7, n)
x2 = np.random.randint(1,7, n)
y = x1 + x2

fig, ax = plt.subplots(nrows=1, ncols=4, figsize=[15,5])

ax[0].plot(y,'ob')
ax[0].set_xlabel("Realização")
ax[0].set_ylabel("$\Omega$")

histograma_x1 = ax[1].hist(x1, bins=range(1,8), density=True)
ax[1].set_title("Histograma de $X_1$")
histograma_x2 = ax[2].hist(x2, bins=range(1,8), density=True)
ax[2].set_title("Histograma de $X_2$")

histograma_y = ax[3].hist(y, bins=range(min(y), max(y)+1), density=True)
ax[3].set_title("Histograma de $X_1 \\times X_2$")

plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=[10,3])

ax[0].plot(x1, y,'ob')
ax[0].set_xlabel("$X_1$")
ax[0].set_ylabel("$Y$")

ax[1].plot(x2, y,'ob')
ax[1].set_xlabel("$X_2$")
ax[1].set_ylabel("$Y$")
```

### VAs Contínuas: Soma de duas Gaussianas

$$\Omega \in \mathbb{R}$$
$$X_1 \sim \mathcal{N}(1, 0.5)$$
$$X_2 \sim \mathcal{N}(0,0.5)$$
$$Y = 2X_1 + X_2$$

```python
n = 5000 
x1 = np.random.normal(1, 0.5, n)
x2 = np.random.normal(0, 0.5, n)
y = 2*x1 + x2

fig, ax = plt.subplots(nrows=1, ncols=4, figsize=[15,5])

ax[0].plot(y,'ob')
ax[0].set_xlabel("Realização")
ax[0].set_ylabel("$\Omega$")

kde_x1 = stats.kde.gaussian_kde(x1)
it_x1 = np.linspace(min(x1), max(x1), n)
histograma_x1 = ax[1].hist(x1, density=True)
ax[1].set_title("Distribuição de $X_1$")
ax[1].set_xlabel("$\Omega_{X_1}$")
ax[1].plot(it_x1, kde_x1(it_x1))

kde_x2 = stats.kde.gaussian_kde(x2)
it_x2 = np.linspace(min(x2), max(x2), n)
histograma_x2 = ax[2].hist(x2, density=True)
ax[2].set_title("Distribuição de $X_2$")
ax[2].set_xlabel("$\Omega_{X_2}$")
ax[2].plot(it_x2, kde_x2(it_x2))

kde_y = stats.kde.gaussian_kde(y)
it_y = np.linspace(min(y), max(y), n)
ax[3].hist(y, density=True)
ax[3].plot(it_y, kde_y(it_y))
ax[3].set_title("Distribuição de $Y$")
ax[3].set_xlabel("$\Omega_Y$")
ax[3].set_ylabel("Densidade")

plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=[10,3])

ax[0].plot(x1, y,'ob')
ax[0].set_xlabel("$X_1$")
ax[0].set_ylabel("$Y$")

ax[1].plot(x2, y,'ob')
ax[1].set_xlabel("$X_2$")
ax[1].set_ylabel("$Y$")
```
